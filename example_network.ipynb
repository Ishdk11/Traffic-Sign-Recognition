```python
import numpy as np
import os
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from keras.utils import np_utils
from keras.layers import Conv2D, MaxPool2D, Flatten, Dropout, Dense, Input, BatchNormalization
from keras.models import Model
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
X_train = np.load('/content/drive/MyDrive/GTSRB/X_train.npy')
X_test = np.load('/content/drive/MyDrive/GTSRB/X_test.npy')
y_train = np.load('/content/drive/MyDrive/GTSRB/y_train.npy')
y_test = np.load('/content/drive/MyDrive/GTSRB/y_test.npy')
```

```python
print("Shape of X_train : ", X_train.shape)
print("Shape of X_test : ", X_test.shape)
print("Shape of y_train : ", y_train.shape)
print("Shape of y_test : ", y_test.shape)
```

```python
y_train = np_utils.to_categorical(y_train,43)
```

```python
len(y_train[0])
```

```python
y_test = np_utils.to_categorical(y_test,43)
```

```python
BATCH_SIZE = 35289
ROWS = 30
COLUMNS = 30
CHANNELS = 3
NUM_CLASSES = 43

(X_train, X_test) = (X_train/255.0, X_test/255.0)
```

```python
input_shape = (ROWS, COLUMNS, CHANNELS)
```

```python
i = Input(shape=input_shape, name="input_layer")
x = Conv2D(filters=32, kernel_size=(3,3), activation='relu', padding='same', name="Conv0")(i)
x = BatchNormalization(name='BatchNormalization0')(x)
x = Conv2D(filters=32, kernel_size=(3,3), activation='relu',padding='same', name='Conv1')(x)
x = MaxPool2D((2,2), name = 'MaxPool0')(x)

x = Conv2D(filters=64, kernel_size=(3,3), activation='relu',padding='same', name='Conv2')(x)
x = BatchNormalization(name='BatchNormalization1')(x)
x = Conv2D(filters=64, kernel_size=(3,3), activation='relu',padding='same', name='Conv3')(x)
x = MaxPool2D((2,2), name = 'MaxPool1')(x)

x = Flatten(name="Flatten")(x)
x = Dropout(0.2, name="Dropout")(x)
x = Dense(1024, activation='relu', name="Dense0")(x)
output = Dense(43, activation='softmax', name="Output")(x)
```

```python
model = Model(inputs=i, outputs=output, name="Test")
model.summary()
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
EPOCHS = 20
model.fit(X_train, y_train, epochs=EPOCHS, validation_data=(X_test, y_test))
```
