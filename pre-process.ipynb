```python
# Get all images from the folder and put their names in a list
def get_all_files_in_list(folder_path):
  image_paths = os.listdir(folder_path)
  return(image_paths)
```

```python
# All the folders have a .csv file in the end. We need to remove that file, as we are manually setting the ground truth
def remove_csv_file(path_list):
  for path in path_list:
    if '.csv' in path:
      path_list.remove(path)
      return path_list
```

```python
# The list only has individual file names. We need to attach the folder path to make the full path
def get_full_image_path(folder_path, path_list):
  full_path_image_list = []
  for path in path_list:
    full_path = os.path.join(folder_path,path)
    full_path_image_list.append(full_path) 
  return full_path_image_list
```

```python
# Manually setting the ground truth class
def make_gt_array(class_number):
  assert type(class_number) == int
  gt = np.array(class_number)
  return gt
```

```python
# Converting the .ppm images into numpy arrays and put all of them in a list
def convert_to_data(full_path_list, class_number):
  data = []
  gt = make_gt_array(class_number)
  for path in full_path_list:
    image = cv2.imread(path)
    image_np = np.asarray(image)
    image_final = np.resize(image_np,(30,30,3))
    data.append([image_final,gt])
  return data
```

```python
# Coverting the list to a numpy array
def convert_to_np(image_data_list):
  data_np = np.asarray(image_data_list)
  return data_np
```

```python
main_folders_list = os.listdir('/content/Drive/MyDrive/GTSRB/Final_Training/Images')
folders_full_paths = get_full_image_path('/content/Drive/MyDrive/GTSRB/Final_Training/Images',main_folders_list)
```

```python
# Get class number directly from the path
def get_class_num(full_folder_path):
  split = os.path.split(full_folder_path)
  class_num = int(split[1])
  return class_num
```

```python
# Get the numpy array of individual classes
def get_array(folder_path):
  class_number = get_class_num(folder_path)
  print(class_number)
  image_names = get_all_files_in_list(folder_path)
  image_names = remove_csv_file(image_names)
  image_paths = get_full_image_path(folder_path, image_names)
  data_list = convert_to_data(image_paths, class_number)
  data_np = convert_to_np(data_list)
  return data_np
```
